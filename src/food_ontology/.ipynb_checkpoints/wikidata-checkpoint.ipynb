```python
def wikidata_class_qury(p,foods):
    url = 'https://query.wikidata.org/sparql'
    query = f"""
    SELECT ?subclass ?subclassLabel ?class ?classLabel
    WHERE 
    {{
      ?subclass wdt:{p} ?class.
      Filter (?class IN ({foods}) )
      SERVICE wikibase:label {{ bd:serviceParam wikibase:language "en". }}
    }}
    """
    #print(query)
    r = requests.get(url, params = {'format': 'json', 'query': query})
    data = r.json()
    food_record = data["results"]['bindings']
    return food_record
```

```python
def wikidata_qury(p,foods):
    foods = foods.split(",")
    add_str = ""
    for food in foods:
        add_str += f"{{?subclass wdt:{p} {food}}} UNION"
        
    add_str = add_str[:-6]
    url = 'https://query.wikidata.org/sparql'
    query = f"""
    SELECT ?subclass ?subclassLabel
    WHERE 
    {{
      {add_str}
      SERVICE wikibase:label {{ bd:serviceParam wikibase:language "en". }}
    }}
    """
    print(query)
    r = requests.get(url, params = {'format': 'json', 'query': query})
    data = r.json()
    food_record = data["results"]['bindings']
    return food_record
```

```python
def slice_list(l, length = 350):
    l_len = len(l)
    
    if l_len < 350:
        return [l]
    a, b = divmod(l_len,length)
    
    return [ l[i*length:(i+1)*length] for i in range(a+1)]
```

```python
temp = data["results"]["bindings"]
candits = ["wd:"+_["item"]["value"].split("/")[-1] for _ in temp]
foods = ",".join(candits[:100])
```

```python
wikidata_qury("P31",foods)
```

```python
with open("food_class.jl", "w", errors="replace") as f:
    candits = ["wd:Q2095"]
    while candits:
        # new candits used for next iteration
        new_candits = []
        # data write to file
        food_class = []
        
        sliced_candits = slice_list(candits, length = 350)
        
        for sl in sliced_candits:
            
            foods = ",".join(sl)
            temp = wikidata_class_qury("P279",foods)

            food_class += [{candit:_} for _ in temp]
            temp_candits = ["wd:"+_["item"]["value"].split("/")[-1] for _ in temp]

            new_candits += temp_candits

        candits = new_candits

        # write output data food_class to file
        wr = jsonlines.Writer(f)
        wr.write_all(food_class)
        
        break
        if not candits:
            break
```
