```python
def food_self_process(food):
    food = food.replace("-"," ")
    food = ''.join(e for e in food if e.isalnum() or e ==" ")
    #print(food)
    food = double_blank(food)
    #print(food)
    return food
    
def double_blank(food):
    pre = 0
    
    for cur in range(len(food)):
        if food[pre] == " " and food[cur] == " ":
            return double_blank(food[:cur]+food[cur+1:])

        pre = cur
            
    return food

def url2name(url):
    part_url = url.split("Reviews-")[-1]
    part_url = part_url.split("-Los_Angeles_California.html")[0]
    
    name_list = part_url.split("-")
    
    if "or" in name_list[0]:
        name = " ".join(name_list[1:])
    else:
        name = " ".join(name_list)
        
    name = name.replace("_"," ")
    return name
        
```

```python
import jsonlines

with open("food_class.jl",errors="ignore") as f:
    wr = jsonlines.Reader(f)
    
    food_name = set()
    for item in wr:
        id_1 = item["subclass"]["value"].split("/")[-1]
        name1 = item["subclassLabel"]["value"]
        
        id_2 = item["class"]["value"].split("/")[-1]
        name2 =item["classLabel"]["value"]
        if id_1 != name1:
            food_name.add(name1.lower())
            
        if id_2 != name2:
            food_name.add(name2.lower())
            
food_name = list(food_name)
len(food_name)
        
fre_tokens = dict()
for food in food_name:
    tokens = food.split(" ")
    
    for token in tokens:
        fre_tokens[token] = fre_tokens.get(token,0)+1
        
# remove stop word
stopword = nlp.Defaults.stop_words
for key in list(fre_tokens.keys()):
    if key in nlp.Defaults.stop_words:
        del fre_tokens[key]
```

```python
with open("rest_food.json", "r") as f:
    rest_food = json.load(f)
```

```python
def process_food(sent_food, food_name, fre_tokens,rest_name):
    for sent, foods in sent_food:
        new_foods = []
        for food in foods:
            if rltk.levenshtein_similarity(food, rest_name)>0.5:
                continue
            food = food_self_process(food)
            
            if food in food_name:
                new_foods.append(food)
                continue
                
            
            nlp_food = nlp(food)
            if len(nlp_food) == 1:
                continue
            
            start = 0
            for token in nlp_food:
                if token.tag_ not in ["NN", "NNP"] and fre_tokens.get(token.text,0) < 10:
                    start = token.i +1
                
                if token.tag_ in ["NN","NNP"]:
                    break
                    
            food = nlp_food[start:]
            new_foods.append(food.text)
        
        yield sent, new_foods
```

```python
restWithFood = dict()
for url in rest_food:
    rest_name = url2name(url)
    sent_food = rest_food[url]
    
    new_sent_food = list(process_food(sent_food, food_name, fre_tokens, rest_name))
    
    restWithFood[url] = new_sent_food
```

```python
def hybrid_similarity(m ,n):
    similarity = rltk.levenshtein_similarity(m,n)
    
    if similarity > 0.7:
        similarity = 1
        return similarity
    else:
        return similarity
```

```python
from rdflib import Graph, URIRef, Literal, XSD, Namespace, RDF, RDFS,BNode
```

```python
g = Graph()
```

```python
MYNS = Namespace("http://dsci558.org/myfakenamespace#")
WD = Namespace("http://www.wikidata.org/entity/")
g.bind("my_ns", MYNS)
g.bind("wd", WD)
```

```python
import jsonlines
with open("food_class.jl",errors="ignore") as f:
    wr = jsonlines.Reader(f)
    
    food_class_dict = dict()
    for item in wr:
        id_1 = item["subclass"]["value"].split("/")[-1]
        name1 = item["subclassLabel"]["value"]
        
        id_2 = item["class"]["value"].split("/")[-1]
        name2 =item["classLabel"]["value"]
        if id_1 != name1 and id_2 != name2:
            g.add((WD[id_1],RDFS.subClassOf, WD[id_2]))
            food_class_dict[name1.lower()] = WD[id_1]
            food_class_dict[name2.lower()] = WD[id_2]
```

```python
block = {}
stopword = nlp.Defaults.stop_words
for food in food_class_dict:
    tokens = food.split(" ")
    
    for token in tokens:
        if token not in stopword:
            block[token] = block.get(token,[])+[food]
```

```python
has_food = MYNS["hasFood"]
has_sent = MYNS["hasSent"]
count = 0
food_class_key = food_class_dict.keys()
for url in res_foodinfo:
    rest = url.split("/")[-1].split(".")[0]
    line = res_foodinfo[url]
    rest_url = URIRef(url)
    
    for food in line.keys():
        node = BNode()
        
        #food_name = food.replace(" ","_")
        line_food = line[food]
        #food_uri = MYNS[rest+"/"+food_name]
        
        g.add((rest_url, has_food, node))
        g.add((node, RDFS.label, Literal(food,lang="en")))
        g.add((node, RDF.value, Literal(line_food["count"],datatype=XSD.integer)))
        
        #calculate the similarity
        highest = (0,"")
        food_list = food.split(" ")
        
        for token in food_list:
            if token in block:
                for item in block[token]:
                    s1 = item.split(" ")
                    similarity = rltk.hybrid_jaccard_similarity(set(s1), set(food_list),function=hybrid_similarity)
                    
                    if similarity > highest[0]:
                        highest = (similarity, item)
                
        # link to knowledge graph
        if not highest[1]:
            continue
        g.add((node, RDF.type, food_class_dict[highest[1]]))
        
        for sent in line_food["sent"]:
            g.add((node, has_sent, Literal(sent,lang="en")))
    print("\r",count, end = "")
    count += 1
```

```python
g.serialize('food2.ttl', format="turtle")
```
